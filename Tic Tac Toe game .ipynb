```python
def player1(gameboard):
    
    choice = True
    
    def row_choice():
    
        row_choice = ''

        while row_choice not in ['0','1','2'] or row_choice.isdigit() == False:

            row_choice = input("In order to pick a position to replace firstly choose a row (0,1,2): ")

            if row_choice.isdigit() == False:
                print('Sorry, that is not a digit! Please, try again.')

            if row_choice.isdigit() == True:

                if row_choice not in ['0','1','2']:
                    print('Sorry, but you did not choose a valid position (0,1,2). Please, try again.')


        return int(row_choice)
    
    def pos_choice():
    
        pos_choice = ''

        while pos_choice not in ['0','1','2'] or pos_choice.isdigit == False:

            pos_choice = input("Choose a position to replace (0,1,2): ")

            if pos_choice.isdigit() == False:
                print('Sorry, that is not a digit! Please, try again')

            if pos_choice.isdigit() == True: 

                if pos_choice not in ['0','1','2']:
                    print('Sorry, but you did not choose a valid position (0,1,2). Please, try again')

        return int(pos_choice)
    
    def replacement_choice1(gameboard,row_choice,pos_choice):
    
        gameboard[row_choice][pos_choice] = 'x'
    
        return gameboard
    
    while choice:
        
        row = row_choice()
        pos = pos_choice()
        
        if gameboard[row][pos] == 'o':
            
            print("The spot is already taken by 'o'! Pick another one.")
            continue
        
        if gameboard[row][pos] == 'x':
            
            print("The spot is already taken by 'x'! Pick another one.")
            continue
        
        else:
            
            gameboard = replacement_choice1(gameboard,row,pos)
            choice = False
    
    return gameboard
```

```python
def win_condition(gameboard):
    
    l = len(gameboard[0])
    win = False
    
    if list(gameboard[i][i] for i in range(0,l)) == ['x','x','x'] or list(gameboard[l-1-i][i] for i in range(l-1,-1,-1)) == ['x','x','x']:
        
        print('Player 1 wins! The game is over.')
        win = True
    
    elif list(gameboard[i][i] for i in range(0,l)) == ['o','o','o'] or list(gameboard[l-1-i][i] for i in range(l-1,-1,-1)) == ['o','o','o']:
        
        print('Player 2 wins! The game is over.')
        win = True
    
    for k in range(len(gameboard)): 
        
        if gameboard[k][:3] == ['x','x','x']:
            print('Player 1 wins! The game is over.')
            win = True
        elif gameboard[k][:3] == ['o','o','o']:
            print('Player 2 wins! The game is over.')
            win = True
        elif coloumn_matrix(gameboard,k) == ['x','x','x']:
            print('Player 1 wins! The game is over.')
            win = True
        elif coloumn_matrix(gameboard,k) == ['o','o','o']:
            print('Player 2 wins! The game is over.')
            win = True
            
    if win == True:
        return True
    else:
        return False
```

```python
def game_on_choice():
    
    choice = 'wrong'
    
    while choice not in ['Y','N','y','n']:
        
        choice = input('Would you like to play again? Y or N ')
        
        if choice not in ['Y','N','y','n']:
            
            clear_output()
            print('Incorrect input value. Please make sure to choose Y or N.')
            
    if choice == 'Y' or choice == 'y':
        return False
    else:
        print('The game is over.')
        return True
```

```python
def quit_game():
    
    choice = ''
    
    while choice not in ['Y','N','y','n']:
        
        choice = input('Would you like to stop playing? Y or N ')
        
        if choice not in ['Y','N','y','n']:
            
            clear_output()
            print('Incorrect input value. Please make sure to choose Y or N.')
    
    if choice == 'Y' or choice == 'y':
        return True
    else:
        return False
```

```python
def full_board(gameboard):
    
    for i in range(len(gameboard[0])):
        if ' ' in gameboard[i]:
            return False
    else:
        return True
```
